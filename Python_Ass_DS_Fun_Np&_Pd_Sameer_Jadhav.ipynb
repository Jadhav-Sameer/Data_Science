1. Create a list of 5 integers.
Perform and print the result of the following operations:
append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
# Create a list of 5 integers
numbers = [10, 20, 30, 40, 50]
print("Original List:", numbers)
```

```python
# 1. append() : Add a single element at the end
numbers.append(60)
print("After append(60):", numbers)
```

```python
# 2. extend() : Add multiple elements at the end
numbers.extend([70, 80])
print("After extend([70, 80]):", numbers)
```

```python
# 3. insert() : Insert at a specific position (index, value)
numbers.insert(2, 25)  # Insert 25 at index 2
print("After insert(2, 25):", numbers)
```

```python
# 4. remove() : Remove first occurrence of a value
numbers.remove(40)  # Removes number 40
print("After remove(40):", numbers)
```

```python
# 5. pop() : Remove element at a given index (default last)
popped = numbers.pop()  # Removes last element
print("After pop():", numbers, "| Popped value:", popped)
```

```python
# 6. index() : Get index of first occurrence
print("Index of 30:", numbers.index(30))
```

```python
# 7. count() : Count occurrences of an element
print("Count of 20:", numbers.count(20))
```

```python
# 8. sort() : Sort in ascending order
numbers.sort()
print("After sort():", numbers)
```

```python
# 9. reverse() : Reverse the list
numbers.reverse()
print("After reverse():", numbers)
```

```python
# 10. clear() : Remove all elements
temp_list = numbers.copy()   # Copy before clearing to show effect
numbers.clear()
print("After clear():", numbers)
```

- 75–89: 'B'
- 50–74: 'C'

```python
# Taking input from user
marks = int(input("Enter your marks: "))

# Checking conditions
if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 50:
    print("Grade: C")
else:
    print("Grade: Fail")
```

7. Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

```python
# Taking number as input
num = int(input("Enter a number: "))

# Checks if the number is positive
if num > 0:
    print("The number is positive.")
    
    # Checks if it is also even
    if num % 2 == 0:
        print("It is also even.")
    else:
        print("And it is odd.")
elif num == 0:
    print("The number is zero.")
else:
    print("The number is negative.")
```
